```python
from pandas.io.data import DataReader
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
from statsmodels.tsa.ar_model import AR
from statsmodels.graphics import tsaplots
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
```

```python
begin = datetime(2009,1,1)
end = datetime.today()
etrade = DataReader('ETFC','google',begin,end)
```

```python
# non-stationary
plt.style.use('ggplot')
etrade['Close'].plot()
```

```python
# non-stationary (p-value of .74)
adfuller(etrade['Close'])
```

```python
# eliminating trend and seasonality
# taking first difference
# variance different over time
etrade['first_diff'] = etrade['Close'] - etrade['Close'].shift()
etrade['first_diff'].plot()
```

```python
# non-stationary (p-value of .74)
adfuller(etrade['first_diff'])
```

```python
# variance flattens when taking nl
etrade['nl'] = etrade['Close'].apply(lambda x: np.log(x))
etrade['nl'].plot()
```

```python
etrade['orig_var'] = pd.rolling_var(etrade['Close'], 30, min_periods=None, freq=None, center=True)  
etrade['log_var'] = pd.rolling_var(etrade['nl'], 30, min_periods=None, freq=None, center=True)

fig, ax = plt.subplots(2, 1)  
etrade['orig_var'].plot(ax=ax[0], title='Original Variance')  
etrade['log_var'].plot(ax=ax[1], title='Log Variance')  
fig.tight_layout()
```

```python
# more stationary than before
etrade['logged_fd'] = etrade['nl'] - etrade['nl'].shift()  
etrade['logged_fd'].plot()
```

```python
# lags: 1,2,10,30 days
etrade['lag1'] = etrade['logged_fd'].shift()  
etrade['lag2'] = etrade['logged_fd'].shift(2)  
etrade['lag10'] = etrade['logged_fd'].shift(10)  
etrade['lag30'] = etrade['logged_fd'].shift(30)
```

```python
# original vs lagged variable. normally distributed
sns.jointplot('logged_fd','lag1',etrade,kind='reg')
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

acf = plot_acf(etrade['Close'], lags = 20)
plt.title("ACF")
acf.show()
```

```python
pacf = plot_pacf(etrade['Close'], lags = 20)
plt.title("ACF")
pacf.show()
```

```python
# repeating above to find relationships between variable and variable at different lags
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf

lag_corr = acf(etrade['logged_fd'].iloc[1:])
lag_partial_corr = pacf(etrade['logged_fd'].iloc[1:])
```

```python
lg_acf = plot_acf(etrade['logged_fd'], lags = 20)
plt.title("ACF")
lg_acf.show()
```

```python
# no correlation between variable and lag. Random walk.
fig, ax = plt.subplots()  
ax.plot(lag_corr, marker='o', linestyle='--')
```

```python
# decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(etrade['nl'], model='additive',freq=30)
fig = plt.figure()
fig = decomposition.plot()
```

```python
# predicts much smaller variation than actual. reason: random walk. cannot use variable from previous time
# period to predict next day's variable
model = sm.tsa.ARIMA(etrade['logged_fd'].iloc[1:],order=(1,0,0))
results = model.fit(disp=-1)
etrade['forecast'] = results.fittedvalues
etrade[['logged_fd','forecast']].plot()
```

```python
# exponential smoothing model
model = sm.tsa.ARIMA(etrade['logged_fd'].iloc[1:],order=(0,0,1))
results = model.fit(disp=-1)
etrade['forecast'] = results.fittedvalues
etrade[['logged_fd','forecast']].plot()
```

```python
model = sm.tsa.ARIMA(etrade['logged_fd'].iloc[1:],order=(1,1,1))
results = model.fit(disp=-1)
etrade['forecast'] = results.fittedvalues
etrade[['logged_fd','forecast']].plot()
```

```python
results.summary()
```

```python
# Durbin-Watson Statistic. 0-4 range. 2 = no first-order serial corr.
# value < 2: positive serial corr
# value > 2: negative serial corr
sm.stats.durbin_watson(etrade['Close'])
```
